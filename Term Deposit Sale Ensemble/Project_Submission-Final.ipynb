### Encoding

```python
replaceStruct = {
                "education":     {"primary": 1, "secondary": 2 ,"tertiary": 3},
                "month": {"jan": 1, "feb":2 , "mar": 3, "apr": 4,"may": 5,"jun": 6,"jul": 7,"aug": 8,"sep": 9,"oct": 10,"nov": 11,"dec": 12},
                "housing":     {"no": 0, "yes": 1 },
                "default":     {"no": 0, "yes": 1 } ,
                "loan":     {"no": 0, "yes": 1 },
                "Target":     {"no": 0, "yes": 1 }
                    }
oneHotCols=["job","marital","contact","poutcome"]
```
